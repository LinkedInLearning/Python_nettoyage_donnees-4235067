```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler
```

```python
df = pd.read_csv('données_enregistrées.csv')
print("DataFrame original:")
print(df)
```

```python
features = ['Salaire', 'Age', 'Expérience (années)', 'Taille (cm)']
```

```python
normalisation = MinMaxScaler()
df_norm = pd.DataFrame(normalisation.fit_transform(df[features]), columns=[features])
print(df_norm)
```

```python
standardisation = StandardScaler()
df_standard = pd.DataFrame(standardisation.fit_transform(df[features]), columns=[features])
print(df_standard)
```

```python
mise_echelle_robuste = RobustScaler()
df_rob = pd.DataFrame(mise_echelle_robuste.fit_transform(df[features]), columns=[features])
print(df_rob)
```

```python
abs_maximale = MaxAbsScaler()
df_abs = pd.DataFrame(abs_maximale.fit_transform(df[features]), columns=[features])
print(df_abs)
```
