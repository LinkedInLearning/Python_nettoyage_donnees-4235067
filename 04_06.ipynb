```python
import pandas as pd

df = pd.read_csv('data_challenge.csv')

print("DataFrame avec les données pour le défi:")
print(df)
```

```python
import re

ages = df['RawData'].str.extract(r', (\d+),')[0].astype(float)

mean_age = ages.mean()

df['RawData'] = df['RawData'].apply(
    lambda x: re.sub(r', ,', f', {mean_age:.0f},', x) if ', ,' in x else x
)
df
```

```python
def corr_nom_ville(raw_data):
    pattern = r'Berllin|Berl'  
    corr = re.sub(pattern,'Berlin', raw_data)
    return corr

df['RawData'] = df['RawData'].apply(corr_nom_ville)

print("\nDataFrame après la correction des noms des villes:")
print(df)
```

```python
df = df['RawData'].str.split(', ', expand=True)

df.columns = ['Prénom', 'Age', 'Ville', 'NumCarte']
df
```

```python
pattern = r'^\d{13}$'

def check_num_carte_format(NumCarte):
    match = re.match(pattern, NumCarte)
    return bool(match)

df['NumCarteValide'] = df['NumCarte'].apply(check_num_carte_format)
invalides_num = df[df['NumCarteValide'] == False]

print(invalides_num)
```
